### 평점에 따른 추천

```python
movies_df[['title','vote_average','vote_count']].sort_values('vote_average',ascending=False)[:10]
```

가중 평점(Weighted Rationg) = (v/(v+m))*R + (m/(v+m)) * C 
* v 개별영화 투표 횟수 
* m 평점을 부여하기 위한 최소 투표 횟수 
* R 개별 영화에 대한 평균 평점
* C 전체영화에 대한 평균평점 

```python
C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6) #상위 60% 
print('C:',round(C,3), 'm:', round(m,3))
```

```python
def weighted_vote_average(record): 
    v = record['vote_count']
    R = record['vote_average']
    return ( (v/(v+m))*R ) + ( (m/(v+m)) * C )
```

```python
movies_df['weighted_vote'] = movies_df.apply(weighted_vote_average, axis=1)
```

```python
movies_df[['title','vote_average','weighted_vote','vote_count']].sort_values('weighted_vote', ascending=False)[:10]
```

```python
# 추천모델 수정 (유사영화 20개 추출 -> 가중평점 높은 10개 추출 )
def find_sim_movie(df, sorted_ind, title_name, top_n=10) : 
    title_movie = df[df['title'] ==title_name]
    
    title_index = title_movie.index.values
    
    similar_indexes=sorted_ind[title_index, :(top_n*2)]
    similar_indexes=similar_indexes.reshape(-1)
    #기준 영화 인덱스 제외
    similar_indexes=similar_indexes[similar_indexes != title_index]
    
    #top n 2배 해당하는 후보군에서 wieghted_vote가 높은 순으로 top_n만큼 추출  
    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending=False)[:top_n]
```

```python
similiar_moveis = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
similiar_moveis[['title','vote_average','weighted_vote']]
```

## 데이터 가공 및 변환

```python
import pandas as pd 
import numpy as np
```

```python
movies = pd.read_csv('./ml-latest-small/movies.csv')
ratings = pd.read_csv('./ml-latest-small/ratings.csv')
print(movies.shape)
print(ratings.shape)
```

```python
movies.head(2)
```

```python
ratings.head(2)
```

```python
# 사용자별 영화 평점이 나와 있는 행렬로 변경
ratings = ratings[['userId','movieId','rating']]
ratings_matrix = ratings.pivot_table('rating', index='userId',columns='movieId')
ratings_matrix.head(3)
```

```python
#영화ID ->영화이름
rating_movies = pd.merge(ratings, movies, on ='movieId')
ratings_matrix = rating_movies.pivot_table('rating', index='userId',columns='title')

#NaN 값 0으로 치환
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head(3)
```

## 영화간 유사도 산출

```python
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T.head(3)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ratings_matrix_T,ratings_matrix_T)

#item_sim 넘파이 행렬을 Dataframe 으로 전환 
item_sim_df = pd.DataFrame(data=item_sim, index=ratings_matrix.columns, columns=ratings_matrix.columns)
print(item_sim_df.shape)
item_sim_df.head(3)
```

```python
item_sim_df['Godfather, The (1972)'].sort_values(ascending= False)[:6]
```

```python
item_sim_df['Inception (2010)'].sort_values(ascending= False)[1:6]
```

## 아이템기반 KNN CF로 개인화된 추천 구현

```python
#사용자별 예측 평점 
def predict_rating(ratings_arr, item_sim_arr) : 
    ratings_pred = ratings_arr.dot(item_sim_arr) / np.array([np.abs(item_sim_arr).sum(axis=1)])
    return ratings_pred
```

```python
ratings_matrix.head(3)
```

```python
item_sim_df.head(3)
```

```python
ratings_pred = predict_rating(ratings_matrix.values, item_sim_df.values)
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index= ratings_matrix.index, columns= ratings_matrix.columns)
ratings_pred_matrix.head(3)
```

```python
from sklearn.metrics import mean_squared_error

#사용자가 평점을 부여한 영화에 대해서만 MSE를 구함 
def get_mse(pred, actual) : 
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred, actual)

print('아이템 기반 모든 최근접 이웃 MSE: ', get_mse(ratings_pred, ratings_matrix.values))
```

```python
#특정 영화와 가장 비슷한 유사도를 가지고 있는 TOP-N의 유사도만 계산하는 방식으로 변경 
#동일한 식 , top-n만 반영 

def predict_rating_topsim(ratings_arr, item_sim_arr, n=20) : 
    
    #사용자-아이템 평점 행렬 크기만큼 0으로 채운 예측 행렬 초기화 
    pred = np.zeros(ratings_arr.shape)
    
    #사용자-아이템 평점 행렬의 열 크기만큼 루프 수행 
    for col in range(ratings_arr.shape[1]) : 
        #유사도 행렬에서 유사도 큰순으로  nro 데이터 행렬의 인덱스 변환 
        top_n_items = [np.argsort(item_sim_arr[:,col])[:-n-1:-1]]
        #개인화된 예측 평점을 계산 
        for row in range(ratings_arr.shape[0]) : 
            pred[row, col] = item_sim_arr[col, :][top_n_items].dot(ratings_arr[row,:][top_n_items].T)
            pred[row, col] /= np.sum(np.abs(item_sim_arr[col,:][top_n_items]))
    
    return pred
```

```python
np.argsort(item_sim_df.values[:,1])[:-11:-1]
```

```python
ratings_pred = predict_rating_topsim(ratings_matrix.values, item_sim_df.values, n=20)
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index= ratings_matrix.index, columns= ratings_matrix.columns)

print('아이템 기반 TOP-20 최근접 이웃 MSE: ', get_mse(ratings_pred, ratings_matrix.values))
```
